```python
countries = [ 'Kenya', 'USA', 'India','Uganda', 'Nigeria', 'Ghana', 'China']
countries
```

```python
filtered_countries = df2[df2['location'].isin(countries)]
```

```python
filtered_countries = filtered_countries.dropna(subset=['total_cases', 'total_deaths'])
```

```python
filtered_countries.loc[:, ['new_cases', 'new_deaths']] = filtered_countries[['new_cases', 'new_deaths']].fillna(0)
```

```python
filtered_countries
```

```python
filtered_countries.loc[:, ['total_vaccinations', 'people_vaccinated']] = (
    filtered_countries[['total_vaccinations', 'people_vaccinated']]
    .interpolate()
    .fillna(method='bfill')
)
```

```python
import ipywidgets as widgets
from IPython.display import display

filtered_countries['date'] = pd.to_datetime(filtered_countries['date'])

# Dropdown
country_dropdown = widgets.Dropdown(
    options=filtered_countries['location'].unique(),
    description='Country:',
    value='Kenya'
)

# Correctly set date options and index
date_options = sorted(filtered_countries['date'].dt.strftime('%Y-%m-%d').unique())
date_slider = widgets.SelectionRangeSlider(
    options=date_options,
    index=(0, len(date_options)-1),
    description='Date Range',
    orientation='horizontal',
    layout={'width': '700px'}
)

# Update plot function
def update_plot(country, date_range):
    df_country = filtered_countries[
        (filtered_countries['location'] == country) &
        (filtered_countries['date'] >= pd.to_datetime(date_range[0])) &
        (filtered_countries['date'] <= pd.to_datetime(date_range[1]))
    ]
    plt.figure(figsize=(10, 5))
    plt.plot(df_country['date'], df_country['total_cases'], label='Total Cases')
    plt.plot(df_country['date'], df_country['total_deaths'], label='Total Deaths')
    plt.title(f"COVID-19 in {country}")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Display interaction
widgets.interactive(update_plot, country=country_dropdown, date_range=date_slider)
```
